### Data Transformation 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python

df=pd.DataFrame({'a':['one','two','three']*2,'b': [1, 1, 2, 3, 3, 2]})
df
```

```python
df.duplicated() #The method duplicated will check each row at a time and return a bool series
                #that whether the row is duplicated
```

```python
df.drop_duplicates() #It will return a data frame without duplicates
```

```python
df['c']=range(6)
df
```

```python
df.drop_duplicates(['a'])  #Here we are filtering out duplicates on the base of a column
```

```python
data=pd.DataFrame({'Items':['broast','fish','rice','shrimps','steaks','broast','steaks'],
                  'Weight':['2kg','3kg','7kg','1kg','3kg','4kg','9kg']})
```

```python
a=data['Items']
a
```

```python
ctg={'broast':'Chicken','fish':'seafood','shrimps':'seafood','steaks':'chicken','rice':'crop'}
```

```python
data['Categories']=a.map(ctg)  # Here we have mapped the categories section with Items
data
```

```python
data=pd.Series([1,3,-999,4.7,-999,1,4])
data
```

```python
data.replace(-999,np.nan)
```

```python
data.replace({1:'000'}) #The argument can also be dict
```

```python
data=pd.DataFrame(np.arange(12).reshape((3,4)),index=['ohio','viginia','new york'],columns=['one','two','three','four'])
```

```python
data
```

```python
data=data.rename(index=str.title,columns=str.upper)
data
```

```python
data.rename(index={'Ohio':'California'},columns={'THREE':'III'})
```

```python
df=pd.DataFrame({'Ages':[18,20,29,16,19,32,44,56]})
df
```

```python
bins=[0,18,24,34,44,57]
gp=['TEENS','YOUNGS','ADULTS','SENIOR','OLD']
```

```python
df['age group']=pd.cut(df.Ages,bins,labels=gp)
```

```python
df
```
